```python
import pandas as pd
import numpy as np
import sqlite3
from sqlite3 import Error
```

```python
def create_connection(path):
    "function untuk connect ke database (kalau belum bikin db-nya bisa dibuat di sini)"
    connection = sqlite3.connect(path)
    return (connection)

connection = create_connection("db_ng3.sqlite")
```

```python
def execute_query(conn,query):
    "execute query NON SELECT"
    cur = conn.cursor()
    cur.execute(query)
    conn.commit()
```

```python
def select_query(conn,query):
    "execute query SELECT"
    cur = conn.cursor()
    cur.execute(query)
    rows = cur.fetchall()
    
    for row in rows:
        print(row)
```

## Insert all table

```python
assessments_table='''
CREATE TABLE assessments (
    code_module varchar(5),
    code_presentation varchar(5),
    id_assessment INTEGER PRIMARY KEY,
    assessment_type varchar(5),
    date date,
    weight float 
);
'''

execute_query(connection,assessments_table)
```

```python
courses='''
CREATE TABLE courses (
    id_courses INTEGER PRIMARY KEY AUTOINCREMENT,
    code_module varchar(3),
    code_presentation varchar(5),
    module_presentation_length INTEGER
);
'''

execute_query(connection,courses)
```

```python
vle='''
CREATE TABLE vle (
    id_site INTEGER PRIMARY KEY,
    code_module varchar(5),
    code_presentation varchar(5),
    activity_type varchar(15),
    week_from INTEGER,
    week_to INTEGER 
);
'''

execute_query(connection,vle)
```

```python
student_info='''
CREATE TABLE studentInfo (
    code_module varchar(5),
    code_presentation varchar(5),
    id_student INTEGER PRIMARY KEY,
    gender varchar(1),
    region varchar(30),
    highest_education varchar (35),
    imd_band varchar(10),
    age_band varchar(10),
    num_of_prev_attempts INTEGER,
    studied_credits INTEGER,
    disability varchar(1),
    final_result varchar(10)
);
'''

execute_query(connection,student_info)
```

```python
#date dikasih integer karena di spreadsheet bentuknya angka
student_regist='''
CREATE TABLE studentRegistration (
    code_module varchar(5),
    code_presentation varchar(5),
    id_student INTEGER,
    date_registration INTEGER,
    date_unregistration INTEGER,
    FOREIGN KEY (id_student) 
      REFERENCES studentInfo (id_student)
);
'''

execute_query(connection,student_regist)
```

```python
#ID_STUDENT dijadikan primary key karena mencari assessment berdasarkan id student aja. Sengaja ga ditambahkan id_student assesment

student_assessment='''
CREATE TABLE studentAssesment (
    id_assessment INTEGER,
    id_student INTEGER PRIMARY KEY,
    date_submitted INTEGER,
    is_banked INTEGER,
    score INTEGER,
    FOREIGN KEY (id_assessment) REFERENCES assessments (id_assessment)
    FOREIGN KEY (id_student) REFERENCES studentInfo (id_student)
);
'''

execute_query(connection,student_assessment)
```
