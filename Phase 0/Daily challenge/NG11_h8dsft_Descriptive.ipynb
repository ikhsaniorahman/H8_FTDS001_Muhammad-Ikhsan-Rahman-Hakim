```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("covid_19_indonesia_time_series_all.csv")
```

```python
df.head()
```

```python
df.info()
```

### Measure of Central Tendency

```python
#Summary of data distribution
df.describe().T
```

```python
#Modus value
df['New Cases'].mode()[0]
```

### Measure of Variance

```python
#Range of data value
datacols = ['New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 'Total Cases', 'Total Deaths',
           'Total Recovered','Total Active Cases']
for i in datacols:
    print(df[i].name + ' range: ' + str(df[i].max() - df[i].min()))
```

```python
#Quantile/quartile
print(df['New Cases'].quantile([0.25, 0.5, 0.75]))
```

```python
# Plot a box-whisker chart
df['New Cases'].plot(kind='box', title='New Cases Distribution', figsize=(10,8))
plt.show()
```

```python
# Plot a box-whisker chart without outliers
df['New Cases'].plot(kind='box', title='New Cases Distribution', figsize=(10,8), showfliers=False)
plt.show()
```

```python
#New Cases variance and standard deviation
print('Var :',df['New Cases'].var())
print('Std :',df['New Cases'].std())
```
